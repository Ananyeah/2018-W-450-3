```python
# Sampling : Sampling Size Calculator

cursor1 = collection_reference.aggregate([{'$sample': {'size': 16509}}])
```

```python
import random

my_dict = {}
for i in range(16509):
   a = next(cursor1)
   my_dict[i] = a
```

```python
# !pip install folium
import folium

m = folium.Map(location=[ 34.0522,-118.243],
              zoom_start=12)
```

```python
coord = {}
tweets = {}
for i in range(16509):
   coord[i] = dict( (key, value) for (key, value) in my_dict[i].items() if key == "coordinates" and value is not None)
   tweets[i] = dict ((key, value) for (key, value) in my_dict[i].items() if key == "text")
```

```python
k= {k for k, v in coord.items() if v }
k
```

```python
coord[0]['coordinates']['coordinates'][0]
```

```python
folium.Marker(
    location=[coord[0]['coordinates']['coordinates'][1],coord[0]['coordinates']['coordinates'][0]],
    popup=tweets[0]['text'],
    icon=folium.Icon(icon='cloud')
    ).add_to(m)
```

```python
for i in k:
    folium.Marker(
    location=[coord[i]['coordinates']['coordinates'][1],coord[i]['coordinates']['coordinates'][0]],
    popup=tweets[i]['text'],
    icon=folium.Icon(icon='cloud')
    ).add_to(m)
```
